# Grouping and Agg functions

```python
data_groupby_location = data.groupby(by = data.Location).mean()
data_groupby_location
data_groupby_location.sort_values("Rainfall" , ascending=False)

# Location which receives most rainfall = Cairns
 
```

### Hot chocolate is the most sold product in the cold months. Find month which is the coldest so that the inventory team can keep the stock of hot chocolate ready well in advance. 

```python
#groupby month and find the coldest
#data.groupby(by="month").ag

data_groupby_month = data.groupby(by = data.month).mean()
data_groupby_month.sort_values("MinTemp" , ascending=True)

# Month which is coldest= July
```

# Apply custom forumla to groupby

$ WCI = (10 * \sqrt{v} - v + 10.5) .(33 - T_{m}) $

v is the speed of the wind and $ T_{m} $ is the minimum temperature

Add a column for WCI and find the month with the lowest WCI. 

```python
from math import sqrt

def funcWCI(x): 
    v= x.WindGustSpeed
    m=x.MinTemp
    return ((10 * sqrt(v) - v + 10.5)*(33 - (m)))
       
    

data['WCI'] = data.apply(funcWCI,axis=1) 
data.head()
data_groupby_wci = data.groupby(by = data.month).mean()
data_groupby_wci.sort_values("WCI")


#Month with lowest WCI = Jan
```

## read data from junesales.csv and merge with weatherdata.csv
#### total tea for the Newcastle store for the month of June 2011 is given in the file names `junesales.csv`
#### merge the Newcastle store data for june month with tea sales on given dates with weather data

```python
junesales = pd.read_csv("junesales.csv")
junesales["day"] = pd.DatetimeIndex(junesales.Date).day
junesales.head()
junesales.shape
```

```python
#Merge columns = localtion =Newcastle  ,  month =6 , dayOfMonth = 1,,3,4,5
newCastle_data = data[((data.Location == "Newcastle") & (data.month==6) & (data.Year==2011))]
mergedata = newCastle_data.merge(junesales , on ="day")
mergedata.columns
```

#### Types of joins. 

* LEFT JOIN  : Keep all data in left, merge data from right on only those which match

* RIGHT JOIN : Keep all data in right, merge data from left on only those which match

* FULL JOIN : union of both data

* INNER JOIN : Intersection of both data

## Pivot table

#### Using pivot tables find the average monthly rainfall in the year 2016 of all the locations. The information can then be used to predict the sales of tea in the year 2017.

```python
# avg monthly raifall in 2016
data_2016 = data[(data.Year==2016)]
#groupby can be only done on 1 column and agg func applies to all the columns
#eg, groupby location, will give agv rainfall for all year not per month
# use pivot table to groupby location and also month for interested values, ie rainfall
data_2016.pivot_table(index = "Location", columns = "month", values =["Rainfall", "WCI"], aggfunc='mean')

```
