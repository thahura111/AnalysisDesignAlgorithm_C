```python
#Investigate the variation of prices of diamonds and note your observations. 


# Before plotting extract the required data by performing necessary data manipulation
groupdata = data.groupby("cut").count()
groupdata

groupdata["price"]




```

```python
list_category = []
list_price = []

for i in [0,1,2,3,4] :
    list_category.append( groupdata.index[i]) 
    list_price.append( groupdata.iloc[i,0]  )#row , column

print(list_category)
print(list_price)
```

```python
#Perform a similar variation analysis on the quality of cut. 

#Investigate the variation of prices of diamonds and note your observations. 
plt.bar(list_category , list_price)

# Adding and formatting title
plt.title("Variation of diamond cut vs Price \n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("cut", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("price", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'} )

plt.show()
```
