#### Cricket Analysis
#### The strike rate of a batsman is defined as the (number of runs scored / number of 100 balls faced). It is a measure of how fast a batsman has played.For example, if a batsman scores 40 runs in 50 balls, the strike rate is 80 runs/100 balls.
#### Among all the innings where a century was scored by a batsman, which one has the highest strike rate?

```python
odi_with_centuries = odi[ (odi.Centuries>=1)]
odi_with_centuries.head()
```

```python
strikerate=  lambda runs,balls : (runs*100)/balls 
odi_with_centuries["StrikeRate"] =odi_with_centuries.apply(lambda x : strikerate(x.Runs, x.Balls),axis=1 )
player= odi_with_centuries[(odi_with_centuries.StrikeRate == odi_with_centuries.StrikeRate.max())]
player
```

### Extract all the relevant metrics (such as the date, day, month, year, hour, and minutes) of each persons

```python
import pandas as pd
grades = pd.read_csv('grades.csv')
grades.head()
```

```python
grades["day"] = pd.DatetimeIndex(grades.submit_time).day
grades["month"] = pd.DatetimeIndex(grades.submit_time).month
grades["year"] = pd.DatetimeIndex(grades.submit_time).year
grades["time"] = pd.DatetimeIndex(grades.submit_time).time
grades["date"]  = grades.submit_time.apply( lambda x : x.split("-")[0])
grades.head()
```

```python
def get_hr_min_sec(x,ip):
    x = str(x)
    y=x.split(':')
    if(ip>=0 and ip <=2):
        return y[ip]
    else:
        return -1
    

grades["hour"] = grades.time.apply(lambda x : get_hr_min_sec(x,0))
grades["minute"] = grades.time.apply(lambda x : get_hr_min_sec(x,1))
grades["second"] = grades.time.apply(lambda x : get_hr_min_sec(x,2))
grades.head()
```

```python
grades["format"] = grades.submission.apply(lambda x : x.split('.')[-1])

sub_zip = grades[(grades["format"]=='zip')]
sub_zip.format.count()  #189
#grades.format.count()  #198
(sub_zip.format.count()/grades.format.count())*100
```

```python
grades.day.dtype
```

```python
### How many students submitted the assignment after the first deadline (including the students who submitted after the second deadline) ?
### first deadline = Jan 3, 2017 - 11:59:59
afterFirstDeadline = grades[ ( (grades.year>2017) | ((grades.year==2017) & (grades.month>1) ) | ( (grades.year == 2017) &  (grades.month==1)  & (grades.day>3) ) ) ]
afterFirstDeadline.head()
afterFirstDeadline.describe()
#ans=44
```
